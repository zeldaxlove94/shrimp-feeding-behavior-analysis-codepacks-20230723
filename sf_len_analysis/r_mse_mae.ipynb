```python
import numpy as np

r_mse_mae = np.genfromtxt('r_mse_mae.csv', delimiter=',')

# Read Data

r = r_mse_mae[:, 0] 
mse = r_mse_mae[:, 1] 
mae = r_mse_mae[:, 2] 

# print list

print("a:", r)
print("b", mse)
print("c", mae)
```

```python
avg_r = np.mean(r)
avg_mse = np.mean(mse)
avg_mae = np.mean(mae)

print("avg_r:", avg_r)
print("avg_mse", avg_mse)
print("avg_mae", avg_mae)
```

```python
import matplotlib.pyplot as plt

plt.hist(r, bins=10, rwidth=0.5,color='r')

plt.xlabel('R-squared')
plt.ylabel('quantity')
plt.title('Distribution of R-squared')
plt.show()
```

```python
plt.hist(r, bins=10, rwidth=0.5,color='g')

plt.xlabel('MSE')
plt.ylabel('quantity')
plt.title('Distribution of MSE')
plt.show()
```
